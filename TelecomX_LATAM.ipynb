#ðŸ“Œ ExtracciÃ³n

```python
import requests
import pandas as pd

url = "https://raw.githubusercontent.com/sthemonica/alura-voz/refs/heads/main/Dados/Telco-Customer-Churn.json"

solicitud = requests.get(url)
data = solicitud.json()

df = pd.DataFrame(data)

df
```

#ðŸ”§ TransformaciÃ³n

```python
import pandas as pd
df_normalizado = pd.json_normalize(data )
df_normalizado
```

Revisar la estructura del DataFrame

```python
df_normalizado.info()
```

Verificar valores nulos

```python
df_normalizado.isnull().sum()
```

Revisar duplicados (por ID de cliente)


```python
df_normalizado.duplicated(subset='customerID').sum()
```

 Convertir columnas numÃ©ricas al tipo correcto

```python
df_normalizado['account.Charges.Total'] = pd.to_numeric(df_normalizado['account.Charges.Total'], errors='coerce')
df_normalizado['account.Charges.Monthly'] = pd.to_numeric(df_normalizado['account.Charges.Monthly'], errors='coerce')
```

```python
df_normalizado.isnull().sum()
```

```python
df_normalizado = df_normalizado.dropna(subset=['account.Charges.Total'])
```

Unificar categorÃ­as

```python
df_normalizado = df_normalizado[df_normalizado['Churn'].isin(['Yes', 'No'])]
```

Ver valores Ãºnicos en variables categÃ³ricas

```python
df_normalizado['Churn'].unique()
df_normalizado['account.PaymentMethod'].unique()
df_normalizado['customer.gender'].unique()
```

```python
df_normalizado['Churn'] = df_normalizado['Churn'].astype('category')
df_normalizado['account.Contract'] = df_normalizado['account.Contract'].astype('category')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# GrÃ¡fico de barras
sns.countplot(data=df_normalizado, x='Churn', palette='pastel')
plt.title("DistribuciÃ³n de EvasiÃ³n de Clientes")
plt.xlabel("Â¿Cliente se dio de baja?")
plt.ylabel("Cantidad de Clientes")
plt.show()
```

```python
# Pie chart
labels = df_normalizado['Churn'].value_counts().index
sizes = df_normalizado['Churn'].value_counts().values

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.title("ProporciÃ³n de Clientes que se Dieron de Baja vs. Retenidos")
plt.axis('equal')  # Para que el cÃ­rculo estÃ© proporcionado
plt.show()
```

73.4% de los clientes permanecen activos con la empresa.

26.6% de los clientes se han dado de baja.

Esta proporciÃ³n revela que aproximadamente 1 de cada 4 clientes cancela el servicio, lo cual representa una tasa de evasiÃ³n moderadamente alta. Esta situaciÃ³n podrÃ­a tener un impacto negativo en la rentabilidad y sostenibilidad del negocio, especialmente si no se identifican y abordan las causas detrÃ¡s de estas cancelaciones.

**Recuento de evasiÃ³n por variables categÃ³ricas**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Lista de variables categÃ³ricas a analizar
categorical_vars = [
    'customer.gender',
    'account.Contract',
    'account.PaymentMethod',
    'internet.InternetService',
    'phone.MultipleLines',
    'customer.SeniorCitizen'
]

# Crear grÃ¡ficos uno por uno
for var in categorical_vars:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df_normalizado, x=var, hue='Churn', palette='pastel')
    plt.title(f'EvasiÃ³n de Clientes segÃºn {var}')
    plt.xticks(rotation=45)
    plt.xlabel(var)
    plt.ylabel('Cantidad de Clientes')
    plt.legend(title='Churn')
    plt.tight_layout()
    plt.show()
```

 1. GÃ©nero del cliente (customer.gender)


*   Hallazgo: La cantidad de hombres y mujeres que se han retirado es  prÃ¡cticamente la misma.
*  InterpretaciÃ³n: Esto sugiere que el gÃ©nero no influye directamente en la decisiÃ³n de abandonar el servicio.

 2. Tipo de contrato (account.Contract)


*   Hallazgo: Los clientes con contrato mensual presentan la mayor tasa de evasiÃ³n (aproximadamente 1500 cancelaciones).
*  InterpretaciÃ³n: Los contratos mensuales, al no tener compromisos a largo plazo, facilitan que los clientes cancelen el servicio en cualquier momento. Por tanto, el tipo de contrato es un fuerte predictor de churn, y la empresa podrÃ­a enfocar sus estrategias de retenciÃ³n en estos usuarios.

 3. MÃ©todo de pago (account.PaymentMethod)


*   Hallazgo: El mÃ©todo "Electronic check" (cheque electrÃ³nico) tiene una alta proporciÃ³n de cancelaciones.
*   InterpretaciÃ³n: Esto podrÃ­a indicar que los clientes que pagan con mÃ©todos menos automatizados (como cheques electrÃ³nicos) tienen una menor fidelidad o enfrentan mÃ¡s fricciones al pagar. Aunque el medio de pago en sÃ­ mismo podrÃ­a no ser la causa. Esto podria ayudar como indicador para retener a esos clientes.


MÃºltiples lÃ­neas telefÃ³nicas (phone.MultipleLines)


*    Hallazgo: Tanto quienes tienen una sola lÃ­nea como quienes tienen mÃºltiples abandonan el servicio.
*   InterpretaciÃ³n: Si las tasas de churn son similares entre ambos grupos, entonces tener mÃºltiples lÃ­neas no parece ser un factor decisivo. Sin embargo, se podria profundizar observando el porcentaje (no solo la cantidad) para ver si hay diferencia relativa.



Ciudadano Senior (customer.SeniorCitizen)
Dato: Esta columna indica si el cliente es adulto mayor:
